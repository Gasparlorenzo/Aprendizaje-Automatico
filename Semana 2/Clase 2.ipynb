```python
# Importamos las librerías necesarias
import pandas as pd
import numpy as np

# Cargamos el archivo CSV desde la ruta correcta
dat_csv = pd.read_csv(
    r'C:\Users\usuario\OneDrive\Desktop\3! Cuatrimestre\03 - Aprendizaje Automático\01 - Ejercicios\empleados.csv',
    encoding="ISO-8859-1"
)

# Seleccionamos las columnas de interés
datos_x = dat_csv['Nombre']
datos_y = dat_csv['Genero']

# Creamos listas vacías
x = []
y = []

# Recorremos la columna "Nombre" y guardamos los valores en la lista x
for i in datos_x:
    x.append(i)

# Recorremos la columna "Genero" y guardamos los valores en la lista y
for j in datos_y:
    y.append(j)

# Mostramos las listas generadas
print(x)
print(y)

# Convertimos la lista x en un array de NumPy
x_array = np.array(x)

# Mostramos el tipo de dato del array
print(type(x_array))

# Mostramos el contenido del array
print(x_array)
```
