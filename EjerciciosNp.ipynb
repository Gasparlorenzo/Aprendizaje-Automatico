```python
import numpy as np
```

```python
array1 = np.zeros(15)
print("Ejercicio 1:")
print(array1)
```

```python
array2 = np.arange(1, 41)
print("Ejercicio 2:")
print(array2)
```

```python
array3 = np.arange(2, 51, 2)
print("Ejercicio 3:")
print(array3)
```

```python
array4 = np.arange(1, 101, 2)
print(array4)
```

```python
matriz5 = np.random.rand(4, 4)
print(matriz5)
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

suma = a + b
print(suma)
```

```python
matrizA = np.random.randint(1, 11, (3, 3))
matrizB = np.random.randint(1, 11, (3, 3))

resultado = np.dot(matrizA, matrizB)

print("Matriz A:\n", matrizA)
print("Matriz B:\n", matrizB)
print("Resultado:\n", resultado)
```

```python
array8 = np.random.randint(1, 101, 15)

print("Array:", array8)
print("Máximo:", array8.max())
print("Mínimo:", array8.min())
print("Promedio:", array8.mean())
```

```python
matriz9 = np.arange(1, 26).reshape(5, 5)

diagonal = np.diag(matriz9)

print("Matriz:\n", matriz9)
print("Diagonal:", diagonal)
```

```python
matriz10 = np.random.randint(1, 21, (5, 5))

suma_filas = matriz10.sum(axis=1)
suma_columnas = matriz10.sum(axis=0)

print("Matriz:\n", matriz10)
print("Suma de filas:", suma_filas)
print("Suma de columnas:", suma_columnas)
```
